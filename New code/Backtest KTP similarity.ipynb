```python
import awswrangler as wr
import boto3
import numpy as np
import pandas as pd
import requests
import json
import ast
import time
```

```python
df_master = pd.read_csv('face_to_default_data_pull_2022_08_19.csv')
df_master.head(10)
```

```python
url = "http://34.101.124.253/v1/models/predict"
distance=[]
probability=[]
issuccess=[]

for i in range(df_master.shape[0]):
    payload = json.dumps({
      "people_img": df_master.iloc[i,2],
      "id_img": df_master.iloc[i,3]
    })
    headers = {
      'accept': 'application/json',
      'ml_token': '1234567890',
      'Content-Type': 'application/json'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    print('loan_id:', int(df_master.iloc[i,0]),'response:', response.text)
    try:
        inference = ast.literal_eval(response.text)
        distance.append(float(inference["distance"]))
        probability.append(float(inference["probability"]))
        issuccess.append('success')
    except:
        distance.append(None)
        probability.append(None)
        issuccess.append(response.text)

df_master["inference_distance"] = distance
df_master["inference_probability"] = probability
df_master["issuccess"] = issuccess
```
